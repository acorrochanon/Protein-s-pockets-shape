## 5 - Filtering the dataset
Requirements:
- Best Fpocket cavity needs to hold a distance smaller to 5Å w.r.t scPDBs 
- Best Fpocket cavity must be ranked 0 or 1. 
- All cavities must have more than 25 points.

````python
# Get filtered cavities, labels, and atom types, we will be plugging into the e3nn. 
filt_cavities, filt_labels, filt_atypes, fp_info_filt, filt_path_order = filterDataset(flatList(fp_coords), flatList(fp_atypes), 
                                         labels, distances, pid_list, druggable_indices, path_order, flatList(fp_info))

# Cleaning the atom types (some contain apostrophes)
for idx1, pocket in enumerate(filt_atypes):
    for idx2, atom_type in enumerate(pocket):
        filt_atypes[idx1][idx2] = ''.join(elem for elem in atom_type if elem.isalnum())
        
# Generate dictionary of the unique atom types we are working with.
dict_atypes = atypesToDict(filt_atypes)

# Compute volume of every cavity that passed the filtering requirements. 
vols = [getPocketVolume(cav) for cav in filt_cavities]

# Removing the fpocket features we are not interested in: pocket score, Mean b-factor, Volume (replaced)
for idx, pocket in enumerate(fp_info_filt):
    for pos in [8, 4, 0]:
        del(pocket[pos])
    pocket.append(vols[idx])
````

### Splitting into train and test sets

````python
cavities_sets, labels_sets, atypes_sets, features_sets = train_test_split(
                                                        list(set([i.split('/')[-3].split('_')[0] for i in filt_path_order])), # Set of protein ids that will be used
                                                        filt_cavities, # Filtered cavities
                                                        filt_labels, # Filtered labels
                                                        filt_atypes, # Clean and filtered atom types
                                                        fp_info_filt, # Clean and filtered fpocket features. 
                                                        split = 0.8)
````

Create train and test dataframes for the logistic regression

````python
train_dict, test_dict = {}, {}
# Crete the keys for the dictionary and set a list as value for each key. 
keys = range(len(features_sets[0][0]))
for i in keys:
    train_dict[i] = []
    test_dict[i] = []

# Fill the train and test dictionaries
for pocket_train in features_sets[0]:
    for idx, feat_train in enumerate(pocket_train):
        train_dict[idx].append(float(feat_train))
                
for pocket_test in features_sets[1]:
    for idx, feat_test in enumerate(pocket_test):
        test_dict[idx].append(float(feat_test))

# Create pandas dfs
train_df = pd.DataFrame(data = train_dict)
stand_train_df = (train_df - train_df.mean()) / train_df.std()

test_df = pd.DataFrame(data = test_dict)
stand_test_df = (test_df - test_df.mean()) / test_df.std()

print(f'Train dataset shape:{train_df.shape}')
print(f'Test dataset shape:{test_df.shape}')
````

#### Logistic regression

````python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

model = LogisticRegression().fit(stand_train_df, labels_sets[0])
y_pred = model.predict(stand_test_df)
y_true = labels_sets[1]

# Metrics
accuracy = model.score(stand_test_df, y_true)
auc_score = roc_auc_score(y_true, y_pred)
print(f'Accuracy: {accuracy:.2f}, AUC: {auc_score:.2f}')
````
