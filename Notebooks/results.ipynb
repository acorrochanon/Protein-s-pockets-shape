```python
import torch
import numpy as np
from sklearn.metrics import roc_curve, roc_auc_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.linear_model import LogisticRegression

def flatList(list_of_lists) -> list:
    if type(list_of_lists[0][0]) == torch.Tensor and len(list_of_lists[0][0]) == 1:
        return [item.item() for sublist in list_of_lists for item in sublist]
    
    else:
        return [item for sublist in list_of_lists for item in sublist]
```

### Logistic regression

```python
import pandas as pd 

train_fp_features = torch.load('data/Features/geo_train_features.pt')
test_fp_features = torch.load('data/Features/geo_test_features.pt')

train_dict, test_dict = {}, {}
# Crete the keys for the dictionary and set a list as value for each key. 
keys = range(len(train_fp_features[0]))
for i in keys:
    train_dict[i] = []
    test_dict[i] = []

# Fill the train and test dictionaries
for pocket_train in train_fp_features:
    for idx, feat_train in enumerate(pocket_train):
        train_dict[idx].append(float(feat_train))
                
for pocket_test in test_fp_features:
    for idx, feat_test in enumerate(pocket_test):
        test_dict[idx].append(float(feat_test))

# Create pandas dfs
train_df = pd.DataFrame(data = train_dict)
stand_train_df = (train_df - train_df.mean()) / train_df.std()

test_df = pd.DataFrame(data = test_dict)
stand_test_df = (test_df - test_df.mean()) / test_df.std()

print(f'Train dataset shape:{train_df.shape}')
print(f'Test dataset shape:{test_df.shape}')
```

```python
train_true = torch.load('data/Labels/train_labels.pt')
test_true = torch.load('data/Labels/test_labels.pt')

model = LogisticRegression().fit(stand_train_df, train_true)
test_pred = model.predict(stand_test_df)

# Predicted probs for the druggable label
test_proba = model.predict_proba(stand_test_df)[:,1]

# Compute AUC and ROC curve scores for test set
lr_auc = roc_auc_score(test_true, test_proba)
lr_fpr, lr_tpr, lr_ = roc_curve(test_true, test_proba)
print(f'Test AUC: {lr_auc:.4f}')
print(f'Test Recall: {recall_score(test_true, test_pred):.4f}')
print(f'Test Precision: {precision_score(test_true, test_pred):.4f}')
print(f'Test f1: {f1_score(test_true, test_pred):.4f}')
print(f'Test CM:\n {confusion_matrix(test_true, test_pred)}')

# Get training metrics 
train_pred = model.predict(stand_train_df)
train_proba = model.predict_proba(stand_train_df)[:,1]
print(f'\nTrain AUC: {roc_auc_score(train_true, train_proba):.4f}')
print(f'Train Recall: {recall_score(train_true, train_pred):.4f}')
print(f'Train Precision: {precision_score(train_true, train_pred):.4f}')
print(f'Train f1: {f1_score(train_true, train_pred):.4f}')
print(f'Train CM:\n {confusion_matrix(train_true, train_pred)}')
```
