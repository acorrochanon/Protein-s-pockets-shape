````python
import os
import seaborn as sns
import tqdm 
from collections import Counter
import matplotlib.pyplot as plt
````

````python
# First position contains the scPDB directory name
paths = [x[0] for x in tqdm(os.walk(os.getcwd()+'/scPDB'))][1:]
````

## Exploratory data analysis

#### Point distribution

````python
protein_ids = [prot.split('/')[-1].split('_')[0] for prot in paths]
cavities = [getCavityInfo(getCavity(protein)) for protein in paths]

print('Number of unique proteins:',len(set(protein_ids)))
print('Number of ligandable cavities:',len(cavities))
points = [cav.shape[0] for cav in cavities]
#print('Shapes:', points)

# Histogram
plt.figure(figsize=(12,5))
sns.histplot(data = points, bins = 25, kde = True)
plt.xlabel('Number of points')
plt.title('Distribution of points in ligandable binding sites')
plt.savefig("point_distribution.jpg", dpi = 300)
plt.show()
````

#### Number of ligandable sites per protein

````python
# num_cavs_per_protein contains a list of number of cavities per protein
num_cavs_per_protein = list(Counter(protein_ids).values())

#Countplot building
plt.figure(figsize=(7,5))
ax = sns.countplot(x = num_cavs_per_protein)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height()/len(num_cavs_per_protein))
    x = p.get_x() + p.get_width() / 2 - 0.05
    y = p.get_y() + p.get_height()
    ax.annotate(percentage, (x, y), size = 12)
plt.title('Number of cavities per protein')
plt.xlabel('Number of cavities')
plt.savefig("num_cavities_per_protein.jpg", dpi = 300)
plt.show()
````
